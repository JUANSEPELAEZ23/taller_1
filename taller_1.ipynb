## Estrategia de análisis

Se propone realizar un análisis bivariado que explore la relación entre los precios y los ratings en función de variables clave como los barrios, el tipo de habitación y la capacidad máxima. El objetivo es identificar patrones que permitan entender cómo estas características influyen en el valor de los alojamientos y su recepción por parte de los huéspedes. Para ello, se calcularán promedios segmentados de precios y calificaciones según cada barrio, tipo de habitación y capacidad. Esto facilitará la comparación de qué categorías de propiedades son más rentables o cuáles reciben mejores evaluaciones, proporcionando una comprensión más profunda del mercado.

A su vez, se llevará a cabo un análisis de correlación. Este enfoque permitirá identificar tanto oportunidades como riesgos en el mercado. Finalmente, se creará un modelo que tiene como objetivo identificar las combinaciones óptimas de características en propiedades de Airbnb que maximicen tanto el precio promedio por noche como el número de reservas

###  Desarrollo de la estrategia

```python
# hacemos cuartiles
q1 = listings_df["price_float"].quantile(0.25)
q3 = listings_df["price_float"].quantile(0.75)
iqr = q3 - q1
```

```python
# Plotting correlation heatmap among review scores
plt.figure(figsize=(10, 8))
sns.heatmap(
    listings_df[["review_scores_rating", "review_scores_accuracy", "review_scores_cleanliness", "review_scores_checkin", "review_scores_communication", "review_scores_location", "review_scores_value"]].corr(),
    vmin=0.5, vmax=1,
    cmap="Blues"
)
plt.title("Correlation among review scores")
plt.show()
```

```python
correlacion =  listings_df[["review_scores_rating", "price_float", "review_scores_cleanliness", "review_scores_checkin", "review_scores_communication", "review_scores_location", "review_scores_value"]].corr()
correlacion["review_scores_location"].sort_values(ascending=False)
```

Análisis por barrio

```python
precio_promedio  = listings_df[["neighbourhood_cleansed","price_float"]].groupby("neighbourhood_cleansed").mean().sort_values(by = "price_float",ascending=False)
precio_promedio
```

```python
# Número de alquileres por barrio
num_alquileres = listings_df.groupby("neighbourhood_cleansed").size().reset_index(name='num_alquileres')
# Combinar ambos DataFrames
relacion_precio_alquileres = pd.merge(precio_promedio, num_alquileres, on="neighbourhood_cleansed")

# Gráfico de dispersión
plt.scatter(relacion_precio_alquileres['num_alquileres'], relacion_precio_alquileres['price_float'])
plt.xlabel('Número de Alquileres')
plt.ylabel('Precio Promedio')
plt.title('Relación entre el Precio Promedio y el Número de Alquileres por Barrio')
plt.show()
relacion_precio_alquileres
```

Análisis por tipo de habitación

```python
# Precio promedio por room_type
precio_promedio_room_type = listings_df.groupby("room_type")["price_float"].mean().reset_index().sort_values(by = "price_float",ascending=False)
precio_promedio_room_type
```

```python
# Número de alquileres por room_type
num_alquileres_room_type = listings_df.groupby("room_type").size().reset_index(name='num_alquileres')

# Combinar los DataFrames
relacion_precio_alquileres_room_type = pd.merge(precio_promedio_room_type, num_alquileres_room_type, on="room_type")

# Gráfico de barras
plt.bar(relacion_precio_alquileres_room_type['room_type'], relacion_precio_alquileres_room_type['price_float'])
plt.xlabel('Tipo de Habitación')
plt.ylabel('Precio Promedio')
plt.title('Precio Promedio por Tipo de Habitación')
plt.show()
```

Ahora por accommodates

```python
# Precio promedio por accommodates
precio_promedio_accommodates = listings_df.groupby("accommodates")["price_float"].mean().reset_index().sort_values(by = "price_float",ascending=False)
precio_promedio_accommodates
```

Vamos a encontrar la mejor combinacion en cuanto al precio y número de reservas respecto a las demás variables que consideramos importantes


```python

grouped_df = listings_df.groupby(['accommodates', 'neighbourhood_cleansed', 'room_type']).agg({
    'price_float': 'mean',        # Precio promedio
    'id': 'count',                # Número de reservas
    'review_scores_rating': 'mean'  # Rating promedio
}).reset_index()

grouped_df.rename(columns={'id': 'num_reservas', 'review_scores_rating': 'rating_promedio'}, inplace=True)


grouped_df['precio_normalizado'] = (grouped_df['price_float'] - grouped_df['price_float'].min()) / (grouped_df['price_float'].max() - grouped_df['price_float'].min())
grouped_df['reservas_normalizadas'] = (grouped_df['num_reservas'] - grouped_df['num_reservas'].min()) / (grouped_df['num_reservas'].max() - grouped_df['num_reservas'].min())
grouped_df['rating_normalizado'] = (grouped_df['rating_promedio'] - grouped_df['rating_promedio'].min()) / (grouped_df['rating_promedio'].max() - grouped_df['rating_promedio'].min())


grouped_df['score'] = grouped_df['precio_normalizado'] + grouped_df['reservas_normalizadas'] + grouped_df['rating_normalizado']


combinacion_optima = grouped_df.sort_values(by='score', ascending=False).head(5)


combinacion_optima_filtrada = combinacion_optima[combinacion_optima['num_reservas'] > 100]

print(combinacion_optima_filtrada)

```

```python

listings_sample_df =  listings_df.loc[listings_df["price_float"] <= (q3 + 1.5 * iqr)].sample(frac=0.1)
listings_sample_df.shape
```

```python

plt.figure(figsize=(10, 8))
sns.scatterplot(
    x=listings_sample_df["price_float"],
    y=listings_sample_df["review_scores_value"]
)
plt.title("Relationship between price and review score value")
plt.grid()
plt.show()
```

```python
# Another common strategy for working with datasets with high variability is by filtering the dataset by representative groups according to the context
# In this sense, let's work only with the neighbourhoods with more listings (pareto analysis)
neighbourhood_frec_cumsum = listings_df["neighbourhood_cleansed"].value_counts(normalize=True).cumsum()
```

```python
# Plotting pareto analysis for neighbourhood frecuency
plt.figure(figsize=(20, 8))
neighbourhood_frec_cumsum.plot(kind="bar", color="steelblue")
plt.title("Pareto analysis for neighbourhood frecuency")
plt.grid(axis="y")
plt.show()
```

```python
most_representative_neighbourhoods = neighbourhood_frec_cumsum.loc[neighbourhood_frec_cumsum < 0.8].index.tolist()
most_representative_neighbourhoods
```

```python
# Plotting price distribution by neighborhood
fig, ax = plt.subplots(1, 1, figsize=(20, 8))
sns.kdeplot(
    data=listings_df.loc[listings_df["price_float"] <= (q3 + 1.5 * iqr)].loc[listings_df["neighbourhood_cleansed"].isin(most_representative_neighbourhoods)],
    x="price_float",
    hue="neighbourhood_cleansed",
    bw_adjust=.3,
    ax=ax
)
for (neighbourhood, color) in zip(most_representative_neighbourhoods, ["steelblue", "orange", "green"]):
    ax.axvline(listings_df.loc[listings_df["price_float"] <= (q3 + 1.5 * iqr)].loc[listings_df["neighbourhood_cleansed"] == neighbourhood, "price_float"].mean(), color=color, linestyle="dashed", linewidth=2, ymax=0.2)
plt.title("Price distribution by neighbourhood (with means)")
plt.show()
```

```python
# Plotting number of reviews distribution by neighbourhood
fig, ax = plt.subplots(1, 1, figsize=(20, 8))
sns.kdeplot(
    data=listings_df.loc[listings_df["price_float"] <= (q3 + 1.5 * iqr)].loc[listings_df["neighbourhood_cleansed"].isin(most_representative_neighbourhoods)],
    x="number_of_reviews",
    hue="neighbourhood_cleansed",
    bw_adjust=.3,
    ax=ax
)
for (neighbourhood, color) in zip(most_representative_neighbourhoods, ["steelblue", "orange", "green"]):
    ax.axvline(listings_df.loc[listings_df["price_float"] <= (q3 + 1.5 * iqr)].loc[listings_df["neighbourhood_cleansed"] == neighbourhood, "number_of_reviews"].mean(), color=color, linestyle="dashed", linewidth=2, ymax=0.2)
plt.title("Price distribution by neighbourhood (with means)")
plt.xlim([0, 200])
plt.show()
```
